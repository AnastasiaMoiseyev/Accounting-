```python
#We pointed out that x and y.
X = cereal_data[["sugars"]]
Y = cereal_data[["calories"]]
```

```python
from sklearn.linear_model import LinearRegression  
linear_regressor = LinearRegression() 

#Linear regression is a basic and commonly used type of predictive analysis.

#Linear regression is a statistical method for modeling relationships between a dependent variable with a given set of independent variables.

#Linear regression uses the relationship between the data points to draw a straight line through all of them. This line can be used to predict future values.

```

```python
linear_regressor.fit(X, Y)
Y_pred = linear_regressor.predict(X)
```

```python
plt.plot(X, Y,'.')
plt.plot(X, Y_pred, color='red')
plt.show()
```

```python
 #In the above figure,
#X-axis = Independent variable

#Y-axis = Output / dependent variable

#Line of regression = Best fit line for a model
```

```python
import numpy as np #imports numpy. to use numpy, use"np". i.e."np.random.randn"
import matplotlib.pyplot as plt #imports matplotlib.pyplot. to use it, use "plt.plot(x,y,'r.')"
```

```python
x = np.linspace(0,10,50) 
# np.linspace is a tool in Python for creating numeric sequences.

#It is somewhat similar to the np.arange function, in that it creates sequences of evenly spaced numbers structured as a np.array

#The linspace function generates linearly spaced vectors.
```

```python
noise = 10 * np.random.randn(50,)
```

```python
y = 3 * x + 7 + noise
```

```python
plt.plot(x,y,'.');
```
